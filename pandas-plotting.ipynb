## Plotting Functions in Pandas

## Lag Plots

http://pandas.pydata.org/pandas-docs/stable/visualization.html#visualization-lag

```python
data = pd.Series(0.1 * np.random.rand(1000) + 0.9 * np.sin(np.linspace(-99 * np.pi, 99 * np.pi, num=1000)))

fig, ax = plt.subplots(figsize=(12, 8))

pd.plotting.lag_plot(data, ax=ax)
```

## Autocorrelation plots

http://pandas.pydata.org/pandas-docs/stable/visualization.html#autocorrelation-plot

```python
fig, ax = plt.subplots(figsize=(12, 8))

data = pd.Series(0.7 * np.random.rand(1000) + 0.3 * np.sin(np.linspace(-9 * np.pi, 9 * np.pi, num=1000)))

pd.plotting.autocorrelation_plot(data, ax=ax)
```
